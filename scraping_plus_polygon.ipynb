```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage containing the table
url = 'http://openinsider.com/screener?s=&o=&pl=&ph=&ll=&lh=&fd=730&fdr=&td=0&tdr=&fdlyl=&fdlyh=&daysago=&xp=1&vl=&vh=&ocl=&och=&sic1=-1&sicl=100&sich=9999&iscfo=1&grp=0&nfl=&nfh=&nil=&nih=&nol=&noh=&v2l=&v2h=&oc2l=&oc2h=&sortcol=0&cnt=800&page=1'

# Send a GET request to fetch the raw HTML content
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
# Locate the table
table = soup.find('table')  # You might need a specific id/class: soup.find('table', {'class': 'example-class'})
```

```python
table = soup.find('table', {'class': 'tinytable'})  # Adjust class or ID if needed
```

```python
table
```

```python
headers = [header.text.strip() for header in table.find_all('th')]

# Extract rows
rows = table.find_all('tr')
data = []
for row in rows[1:]:  # Skip the header row
    cells = row.find_all('td')
    data.append([cell.text.strip() for cell in cells])
```

```python
for i in range(len(headers)):
    headers[i] = headers[i].replace("\xa0", " ")

headers
```

```python
# Convert to a DataFrame
df = pd.DataFrame(data, columns=headers)
```

```python
from datetime import datetime, timedelta

# Imposta i parametri
API_KEY = "F_tdkCtTUqv6OAWg2MG3y65XPu80GyQH"  # Sostituiscilo con la tua API Key
start_date = "2023-12-14"  # Data di inizio nel formato YYYY-MM-DD
num_days = 10  # Numero di giorni

# Calcola la data finale
start_dt = datetime.strptime(start_date, "%Y-%m-%d")
end_dt = start_dt + timedelta(days=num_days)
end_date = end_dt.strftime("%Y-%m-%d")
```

```python
ticker_historical_prices = {}
```

```python
import requests

for ticker in tickers:
    # Costruisci l'URL dell'API
    url = f"https://api.polygon.io/v2/aggs/ticker/{ticker}/range/1/day/{start_date}/{end_date}?apiKey={API_KEY}"

    # Effettua la richiesta
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        if "results" in data:
            ticker_historical_prices[ticker] = pd.DataFrame(data["results"])
        else:
            print("Nessun dato disponibile per l'intervallo specificato.")
    else:
        print(f"Errore: {response.status_code} - {response.text}")
```
