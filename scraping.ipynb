```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage containing the table
url = 'http://openinsider.com/screener?s=&o=&pl=&ph=&ll=&lh=&fd=0&fdr=&td=0&tdr=&fdlyl=&fdlyh=6&daysago=&xp=1&vl=100&vh=&ocl=0&och=&sic1=-1&sicl=100&sich=9999&isofficer=1&iscob=1&isceo=1&ispres=1&iscoo=1&iscfo=1&isgc=1&isvp=1&grp=2&nfl=&nfh=&nil=2&nih=&nol=0&noh=&v2l=&v2h=&oc2l=&oc2h=&sortcol=0&cnt=1000&page=2'

# Send a GET request to fetch the raw HTML content
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
# Locate the table
table = soup.find('table')  # You might need a specific id/class: soup.find('table', {'class': 'example-class'})
```

```python
table = soup.find('table', {'class': 'tinytable'})  # Adjust class or ID if needed
```

```python
table
```

```python
headers = [header.text.strip() for header in table.find_all('th')]

# Extract rows
rows = table.find_all('tr')
data = []
for row in rows[1:]:  # Skip the header row
    cells = row.find_all('td')
    data.append([cell.text.strip() for cell in cells])
```

```python
data
```

```python
headers
```

```python
for i in range(len(headers)):
    headers[i] = headers[i].replace("\xa0", " ")

headers
```

```python
# Convert to a DataFrame
df = pd.DataFrame(data, columns=headers)
```

```python
df
```

```python
df["Company Name"]
```

```python
df = df[~ (df["X"].str.contains("D"))]
```

## Stock values
----------------

```python
df = df.reset_index()
```

```python
tickers = df["Ticker"].unique()
tickers, len(tickers)
```

```python
import yfinance as yf
from datetime import datetime, timedelta
```

```python
stock_data = {}
```

```python
for index, row in df.iterrows():
    filing_date = row["Filing Date"].split(" ")[0]
    stock =  row["Ticker"] 
    date_object = datetime.strptime(filing_date, "%Y-%m-%d")
    start = date_object + timedelta(days=1)
    end = date_object + timedelta(days=180)


    print(f"Downloading data for {stock}...")
    stock_data[stock] = yf.download(stock, start=start, end=end)


    if index == 5: break
```
